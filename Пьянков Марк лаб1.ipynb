```python
# разбейте исходную таблицу на 2: по полу пассажиров
df = pd.read_csv("titanic.csv")
men = df[df['Sex'] == 'male']
women = df[df['Sex'] == 'female']
men
```

```python
pd.set_option('display.max_colwidth', 200)
women
```

```python
# удалите пустые строки или заполните их соответствующими символами
women = women.dropna()
men = men.dropna()
women.info()
```

```python
# Построить гистограмму распределения пассажиров по возрасту
men['Age'].hist()#men['Age'].hist()
plt.show()
women['Age'].hist(color='red')#men['Age'].hist()
```

```python
# Вычислите мин и макс, среднее, моду, медиану возраста пассажира каждого пола
print("MEN")
print("Минимальный возраст", men["Age"].min())
print("Максимальный возраст", men["Age"].max())
print("Средний возраст", men["Age"].mean())
print("Мода(наиболее встречаемый)", men["Age"].mode()[0])
print("Медиана возраста", men["Age"].median())
```

```python
print("WOMEN")
print("Минимальный возраст", women["Age"].min())
print("Максимальный возраст", women["Age"].max())
print("Средний возраст", women["Age"].mean())
print("Мода(наиболее встречаемый)", women["Age"].mode()[0])
print("Медиана возраста", women["Age"].median())
```

```python
# визуализировать возраст при помощи различных настроек метода plot
age = men['Age']
age = age.sort_values()
buckets = [0, 0, 0, 0]

bins = [0, 15, 24, 35, 50, float('inf')]
labels = ['<15', '<24', '<35', '<50', '>50']

group = pd.cut(age, bins=bins, labels=labels, right=False) # ставим метки 

age = df = pd.concat([age, group], axis=1, keys=['Age', 'group']) # соединяем в 1 DF

age = (df.groupby('group')).agg({'Age': 'count'}) # группируем и агреггируем по количеству

age.plot(kind='pie', subplots=True)
age.plot(kind='bar', subplots=True)
```
